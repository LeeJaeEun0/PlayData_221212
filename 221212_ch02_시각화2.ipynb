```python
np.random.seed(1212)
df1 = pd.DataFrame(
    # 나열된 데이터
    # rand, randn, randint, choice...?
    np.random.randn(100, 3), # 100 x 3 행렬 모양으로 정규분포를 따르는 난수 만들기
    # 그만큼의 index
    index=pd.date_range('2022.12.12', periods=100), # 시작점, 갯수
    # 그만큼의 열
    columns=['A', 'B', 'C']
).cumsum() # 0행부터 -> 누적해서 합해주겠다 (cumsum -> cumulative sum : 누적합)
# 누적곱: cumprod()
df1
```

```python
# plot 명령 -> pandas 데이터 시각화
df1.plot()
# 사람이 알아보기 어려움
# 다른 형태의 차트로 바꿀 수 있음
```

```python
# https://pandas.pydata.org/pandas-docs/version/1.3/reference/api/pandas.DataFrame.plot.html
# https://wikidocs.net/159927
df1.plot() # plot 활성화되어 있음 - 판다스에서 plt를 불러서 사용가능
plt.title('dl 내장 plt')
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()
```

```python
df1.plot(figsize=(100, 20)) # plot 활성화되어 있음 - 판다스에서 plt를 불러서 사용가능
plt.title('dl 내장 plt')
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()
```

### 플롯 종류 선택 (`kind`)

```python
iris = sns.load_dataset('iris')# 붓꽃
titanic = sns.load_dataset('titanic')
```

```python
iris
```

```python
iris.columns
```

```python
iris.sepal_length[:20].plot() # 0-19까지 20개의 데이터를 뽑음
```

```python
iris.sepal_length.head(20).plot() # 위와 동일
```

```python
iris.sepal_length[:20].plot(kind='bar') # bar 차트로 변형
```

```python
iris.sepal_length[:20].plot(kind='bar', rot=0) # bar 차트로 변형, 차트 방향 변경
```

```python
# kind = 종류 => 해당 종류 plot 교체
iris.sepal_length[:20].plot(kind='bar', rot=0)
plt.title('꽃받침의 길이 시각화')
plt.xlabel('Data')
plt.ylabel('꽃받침의 길이')
plt.show() # 차트 호출 선언, 마지막값인 20 - 1까지 출력됨!
```

```python
# kind를 사용하지 않고 메소드를 직접 호출 - 해당 차트의 메소드를 호출해서 사용
iris[:5].plot.bar(rot=0) #colum 지정 x, 모든 열에 대한 bar 차트
```

```python
iris.plot.bar(rot=0) #colum 지정 x, 모든 열에 대한 bar 차트
```

```python
# kind를 사용하지 않고 메소드를 직접 호출 - 해당 차트의 메소드를 호출해서 사용
iris[:5].plot.bar(rot=0) #colum 지정 x, 모든 열에 대한 bar 차트
plt.ylim(0,7) #plt에서 최대, 최소값 지정
```

```python
#  파이 차트: 원형, 비율을 나타낼때
df2 = titanic.pclass.value_counts() #고윳값에 대한 빈도수
df2
```

```python
df2.plot.pie()
```

```python
df2.plot.pie(autopct='%.2f%%') # 포맷설정 # 모양이 살짝 일그러짐
```

```python
df2.plot.pie(autopct='%.2f%%') # 포맷설정
plt.axis('equal') # x축, y축 비율
plt.show()
```

```python
# 히스토그램
# 가로축에 분포(값), 세로축에 빈도(몇개...)를 시각화
iris.plot.hist()
```

```python
# 커널 밀도 그래프
# 부드럽게 그린 히스토그램 - 위의 그래프보다 보기 편함
iris.plot.kde()
```

```python
# 박스 플롯 - 5대 요약한 수치 제공 + 이상치 추론
iris.plot.box() # 5대 요약 수치 - 중간값 / 최댓값, 최솟값은 예측한 최대,최소값이다.
# => 일반적인 최대,최소가 아님. 임의로 구한 값(추산)
# 박스 플롯은 5가지 값 = 1,2,3분위 값 + 예상 최댓값과 예상 최솟값
# 동그란 전구?는 예상 값을 벗어난경우? 사용
```

```python
iris.boxplot(by='species')
# padding : 안쪽 여백
plt.tight_layout(pad=3, h_pad=1)
plt.show()
```
